# Time series - Decomposing and differencing

## Importing libraries and dataset

```python
import pandas as pd
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose 
```

```python
df = pd.read_csv("us-airlines-monthly-aircraft-miles-flown.csv", header = 0, parse_dates = [0])
df.head()
```

## Decomposing time series

### additive model
y(t) = Level + Trend + Seasonality + Noise
### Multiplicative model
y(t) = Level * Trend * Seasonality * Noise

```python
df.index = df["Month"]
df.head()
```

```python
result = seasonal_decompose(df["MilesMM"], model = "additive")
result.plot()
```

```python
result2 = seasonal_decompose(df["MilesMM"], model = "multiplicative")
result2.plot()
```

## Defferencing Time series

```python
df["lag1"] = df["MilesMM"].shift(1)
df["MilesMM_diff_1"] = df["MilesMM"].diff(periods = 1)
df.head()
```

```python
result.plot()
```

```python
result3 = seasonal_decompose(df.iloc[1:, 3], model = "additive")
result3.plot()
```

```python
df["MilesMM_diff_12"] = df["MilesMM"].diff(periods = 12)
df.head(13)
```

```python
result4 = seasonal_decompose(df.iloc[12:, 4], model = "additive")
result4.plot()
```
