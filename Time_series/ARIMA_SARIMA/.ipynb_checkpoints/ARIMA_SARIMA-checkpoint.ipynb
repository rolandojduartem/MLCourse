From the graph, We select q = 2 where we have over 95% confidence

## ARIMA(p, d, q) model

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(df["Sales"], order = (5, 2, 2))
model_fit = model.fit()
```

```python
model_fit.summary()
```

```python
resid = model_fit.resid
resid.plot()
```

```python
resid.describe()
```

model = ARIMA(df.Sales, order = (0, d, q)

```python
output = model_fit.forecast()
output
```

### Walk forward ARIMA

```python
df.head()
```

```python
train_size = int(df.shape[0] * 0.7)
train, test = df.Sales[0: train_size], df.Sales[train_size:]
print(train.shape, test.shape)
```

```python
data = train
predict_warima = []
for t in test:
    model = ARIMA(data, order = (5, 2, 2))
    model_fit = model.fit()
    y = model_fit.forecast()
    print(y.values[0])
    predict_warima.append(y.values[0])
    data = np.append(data, t)
    data = pd.Series(data)
```

```python
from sklearn.metrics import mean_squared_error
mse_warima = mean_squared_error(test, predict_warima)
mse_warima
```

```python
plt.plot(test, color = "blue")
plt.plot(test.index, predict_warima, color = "orange")
plt.legend(["Real values", "Predicted values (ARIMA)"])
plt.show()
```

### SARIMA
SARIMA(p, d, q)(P, D, Q)m<br>
- p: Trend autoregression order
- d: Trend difference order
- q: Trend moving average  order<br>
### Seasonal elements:
There are four seasonal elements that are not part  of ARIMA that must be configured , they are:
- P: Seasonal autoregression order
- D: Seasonal difference order
- Q: Seasonal moving average order
- m: The number of time steps for a single seasonal period.

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
df = pd.read_csv("us-airlines-monthly-aircraft-miles-flown.csv", header = 0, parse_dates = [0])
df.head()
```

```python
df.MilesMM.plot()
```

From the graph, we can see seasonality and a increasing linear trend

```python
df.index = df["Month"]
df.head()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result_a = seasonal_decompose(df.MilesMM, model = "multiplicative")
result_a.plot()
```

From the last graph, it is easy to observe an increasing linear trend and seasonality (yearly)

```python
autocorrelation_plot(df.MilesMM)
```

```python
plot_pacf(df.MilesMM, lags = 15)
```

```python
model = SARIMAX(df.MilesMM, order = (5, 1, 1), seasonal_order = (5, 1, 1, 12))
model_fit = model.fit()
```

```python
residuals = model_fit.resid
residuals.plot()
```

```python
output = model_fit.forecast()
output
```

```python
model_fit.forecast(12).plot()
```

```python
yhat = model_fit.predict()
yhat
```

```python
plt.plot(df.MilesMM, color = "blue")
plt.plot(yhat, color = "orange")
plt.legend(["Real values", "Predicted values (SARIMAX)"])
plt.show()
```
