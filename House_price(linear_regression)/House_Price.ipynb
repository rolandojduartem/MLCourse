## Outliers treatment

```python
raw_df.info()
```

```python
max_lim = np.percentile(raw_df.n_hot_rooms, [99])[0] #value in the 99% of total data in the n_hot_rooms column
max_lim
```

```python
raw_df[raw_df.n_hot_rooms > max_lim]
```

```python
raw_df.n_hot_rooms[(raw_df.n_hot_rooms > 3 * max_lim)] = 3 * max_lim
raw_df[raw_df.n_hot_rooms > max_lim]
```

```python
min_lim = np.percentile(raw_df.rainfall, [1])[0]
min_lim
```

```python
raw_df[raw_df["rainfall"] < min_lim]
```

```python
raw_df.rainfall[raw_df["rainfall"] < 0.3 * min_lim] = 0.3 * min_lim
raw_df[raw_df["rainfall"] < min_lim]
```

```python
raw_df.info()
```

## Missing values treatment

In m_hos_beds there are some missing values

```python
raw_df[raw_df.n_hos_beds.isna()]
```

```python
raw_df.n_hos_beds = raw_df.n_hos_beds.fillna(raw_df.n_hos_beds.mean())
#raw_df = raw_df.fillna(raw_df.mean)
#Also, It can be used raw_df = raw_df.fillna(raw_df.mean)
```

```python
raw_df.info() #There re no more missing values in n_hos_beds column
```

```python
df = raw_df #df is clean data (missing values and outliers are fixed)
```
